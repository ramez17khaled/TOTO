```python
import pandas as pd
```

```python
Car_path = "D:/data/MSDial_Untargeted/MSDial/01-mzmlThermo/RT ok/DB et results NCE/NCE_Car.xlsx"
CE_path = "D:/data/MSDial_Untargeted/MSDial/01-mzmlThermo/RT ok/DB et results NCE/NCE_CE.xlsx"
CER_SM_path = "D:/data/MSDial_Untargeted/MSDial/01-mzmlThermo/RT ok/DB et results NCE/NCE_Cer_SM.xlsx"
DG_path = "D:/data/MSDial_Untargeted/MSDial/01-mzmlThermo/RT ok/DB et results NCE/NCE_DG.xlsx"
LPL_path = "D:/data/MSDial_Untargeted/MSDial/01-mzmlThermo/RT ok/DB et results NCE/NCE_LPL.xlsx"
PC_PE_path = "D:/data/MSDial_Untargeted/MSDial/01-mzmlThermo/RT ok/DB et results NCE/NCE_PC_PE.xlsx"
PG_path = "D:/data/MSDial_Untargeted/MSDial/01-mzmlThermo/RT ok/DB et results NCE/NCE_PG.xlsx"
PI_path = "D:/data/MSDial_Untargeted/MSDial/01-mzmlThermo/RT ok/DB et results NCE/NCE_PI.xlsx" 
PS_path = "D:/data/MSDial_Untargeted/MSDial/01-mzmlThermo/RT ok/DB et results NCE/NCE_PS.xlsx" 
TG_path = "D:/data/MSDial_Untargeted/MSDial/01-mzmlThermo/RT ok/DB et results NCE/NCE_TG.xlsx" 
```

```python
car_data = pd.read_excel(Car_path)
CE_data = pd.read_excel(CE_path)
CER_SM_data = pd.read_excel(CER_SM_path)
DG_data = pd.read_excel(DG_path)
LPL_data = pd.read_excel(LPL_path)
PC_PE_data = pd.read_excel(PC_PE_path)
PG_data = pd.read_excel(PG_path)
PI_data = pd.read_excel(PI_path)
PS_data = pd.read_excel(PS_path)
TG_data = pd.read_excel(TG_path)
car_data
```

```python
Car_mod = car_data
Car_mod = Car_mod.groupby(["Compound", "Peak Label"]).first().reset_index()
Car_mod['Charge'] = 'Positive'
Car_mod['Family'] = 'Car'
Car_mod['Adduct'] = '[M+H]+'

CE_mod = CE_data
CE_mod = CE_mod.groupby(["Compound", "Peak Label"]).first().reset_index()
CE_mod['Charge'] = 'Positive'
CE_mod['Family'] = CE_mod['Compound'].apply(
        lambda x: 'CE' if x.startswith('CE') else
                  'Cholesterol' if x.startswith('Cholesterol') else
                  'Unknown')
CE_mod['Adduct'] = '[M+NH4]+'

CER_SM_mod = CER_SM_data
CER_SM_mod = CER_SM_mod.groupby(["Compound", "Peak Label"]).first().reset_index()
CER_SM_mod['Charge'] = 'Positive'
CER_SM_mod['Adduct'] = '[M+H]+'
CER_SM_mod['Family'] = CER_SM_mod['Compound'].apply(
        lambda x: 'CER' if x.startswith('Cer') or x.startswith('GM') else
                  'SM' if x.startswith('SM') else
                  'HexCer' if x.startswith('Hex') else
                  'Sphinganine' if x.startswith('Sphinganine') else
                  'Sphingosine' if x.startswith ('Sphingosine') else
                  'Unknown')

DG_mod = DG_data
DG_mod = DG_mod.groupby(["Compound", "Peak Label"]).first().reset_index()
DG_mod['Charge'] = 'Positive'
DG_mod['Family'] = 'DG'
DG_mod['Adduct'] = '[M+NH4]+'

LPL_mod = LPL_data
LPL_mod = LPL_mod.groupby(["Compound", "Peak Label"]).first().reset_index()
LPL_mod['Charge'] = 'Positive'
LPL_mod['Family'] = 'LPL'
LPL_mod['Adduct'] = '[M+H]+'

PC_PE_mod = PC_PE_data
PC_PE_mod = PC_PE_mod.groupby(["Compound", "Peak Label"]).first().reset_index()
PC_PE_mod['Charge'] = 'Positive'
PC_PE_mod['Adduct'] = '[M+H]+'
PC_PE_mod['Family'] = PC_PE_mod['Compound'].apply(
        lambda x: 'PC' if x.startswith('PC') else
                  'PE' if x.startswith('PE') else
                  'Unknown')

PG_mod = PG_data
PG_mod = PG_mod.groupby(["Compound", "Peak Label"]).first().reset_index()
PG_mod['Charge'] = 'Negaitive'
PG_mod['Family'] = 'PG'
PG_mod['Adduct'] = '[M-H]-'

PI_mod = PI_data
PI_mod = PI_mod.groupby(["Compound", "Peak Label"]).first().reset_index()
PI_mod['Charge'] = 'Negaitive'
PI_mod['Family'] = 'PI'
PI_mod['Adduct'] = '[M-H]-'

PS_mod = PS_data
PS_mod = PS_mod.groupby(["Compound", "Peak Label"]).first().reset_index()
PS_mod['Charge'] = 'Negaitive'
PS_mod['Family'] = 'PS'
PS_mod['Adduct'] = '[M-H]-'

TG_mod = TG_data
TG_mod = TG_mod.groupby(["Compound", "Peak Label"]).first().reset_index()
TG_mod['Charge'] = 'Positive'
TG_mod['Family'] = 'TG'
TG_mod['Adduct'] = '[M+NH4]+'


PC_PE_mod
```

```python
dfs = [Car_mod, CE_mod, CER_SM_mod, DG_mod, LPL_mod, PC_PE_mod, PG_mod, PI_mod, PS_mod, TG_mod]

output_lines = []

for df in dfs:
    main_entries = df[~df['Peak Label'].str.contains('F', na=False)]

    for _, main_row in main_entries.iterrows():
        compound_name = main_row['Compound']
        precursor_mz = main_row['m/z (Expected)']
        formula = main_row['Formula']
        retention_time = main_row['RT']
        ion_mode = main_row['Charge']
        compound_class = main_row['Family']
        adduct = main_row['Adduct']

        # Set RETENTIONTIME to empty if retention_time is NaN or 0
        retention_time_str = f"RETENTIONTIME: {retention_time}" if pd.notna(retention_time) and retention_time != 0 else "RETENTIONTIME: "

        num_peaks_rows = df[(df['Compound'] == compound_name) & (df['Peak Label'].str.contains('F', na=False))]

        num_peaks = []
        for _, peak_row in num_peaks_rows.iterrows():
            num_peaks.append(f"{peak_row['m/z (Expected)']:.5f}\t999")
        num_peaks_str = "\n".join(num_peaks)

        entry = f"""NAME: {compound_name}
PRECURSORMZ: {precursor_mz:.5f}
PRECURSORTYPE: {adduct}
SMILES: 
INCHIKEY: 
FORMULA: {formula}
{retention_time_str}
CCS: 
IONMODE: {ion_mode}
COMPOUNDCLASS: {compound_class}
Comment: theoretical MS2 created from the information of Orbitrap Lipidomics.
Num Peaks: {len(num_peaks)}
{num_peaks_str}
"""
        output_lines.append(entry)

with open("HomeDB_MS2.msp", "w") as f:
    f.write("\n".join(output_lines))

print("Data successfully written to HomeDB_MS2.msp.")
```
